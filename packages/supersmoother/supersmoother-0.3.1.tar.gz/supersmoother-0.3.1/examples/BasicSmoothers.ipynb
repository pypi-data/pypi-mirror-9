# Neighborhood-Based Smoothing

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# use seaborn-style plots. This can be safely commented-out.
import seaborn; seaborn.set()
```

```python
rng = np.random.RandomState(0)
x = np.sort(10 * rng.rand(100))
y = np.sin(2 * x) + 0.4 * rng.randn(100)
plt.plot(x, y, '.');
```

# ``MovingAverageSmoother``

The ``MovingAverageSmoother`` is the simplest of the span-based smoothers.
At each point, the smoothed value consists of the mean of nearby values.

```python
from supersmoother import MovingAverageSmoother
model = MovingAverageSmoother(0.1)
model.fit(x, y)

xfit = np.linspace(0, 10, 1000)
plt.plot(x, y, '.')
plt.plot(xfit, model.predict(xfit), '-');
```

```python
plt.plot(x, y, '.')

for span in [0.05, 0.2, 0.5]:
    model = MovingAverageSmoother(span)
    model.fit(x, y)
    plt.plot(x, model.predict(x),
             label="{0:.2f}".format(span))
plt.legend();
```

## ``LinearSmoother``

```python
from supersmoother import LinearSmoother

plt.plot(x, y, '.')

for span in [0.05, 0.2, 0.5]:
    model = LinearSmoother(span)
    model.fit(x, y)
    plt.plot(x, model.predict(x),
             label="{0:.2f}".format(span))
plt.legend();
```
