```python
nt=300
x=np.arange(nt)
```

```python
for stride in [1, 2, 3, 4, 5, 27]:
    for lag in [0, 1, 2, 7, 23]:
        print "lag", lag
        print "stride", stride
        np.testing.assert_equal( x[lag::stride], x[np.arange(lag, nt, stride)] ) 
        wanted = np.arange(lag, nt, stride)
       
        print wanted + lag
```
